## 此為`博客來暢銷書`的爬蟲


```python
#爬蟲

import urllib.request as req
import requests

url= "https://www.books.com.tw/web/sys_saletopb/books/?attribute=7&loc=P_0002_001" 
# 建立一個 Request 物件，附加 Request Headers的資訊
request=req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
})
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")    
print(data)
```

```python
#此網站的標題

import bs4 #beautifulsoup4
root=bs4.BeautifulSoup(data, "html.parser") #用html做解析
print(root.title.string)
```

```python
#試著抓取所需資料部分

titles=root.find_all("div", "type02_bd-a")# 尋找 class="type02_bd-a" 的 div 標籤
print(titles)
```

```python
#印出一個書名

books=root.find("h4").string #使用 root.find取出資料
print(books)
```

```python
#印出全部書的資料

for item in titles:
    if item.find('h4').string != None :#如果資料非空白就設定變數
        books = item.find('h4').string
    if item.find('li').a != None :
        autor = item.find('li').a.string
    if item.find('li', 'price_a').b != None :
        money = item.find('li', 'price_a').text  #使用text取得字串

        print('書名：', books,'，作者：',autor,'，',money) #印出看看 
```

```python
#存為csv檔

import csv
data = root.find_all("div", "type02_bd-a")
```

```python
# 使用 with open            參考網站：https://blog.gtwang.org/programming/python-csv-file-reading-and-writing-tutorial/

# 開啟 CSV 檔案             newline='' 參數是為了讓資料中包含的換行字元可以正確被解析
with open('data.csv', 'w', newline='', encoding='UTF-8') as csvfile: #傳入識別符號 W，表示寫檔    
    
    writer = csv.writer(csvfile)                          # 建立 CSV 檔寫入器  
    writer.writerow(['書名', '作者', '售價'])              # 寫入一列資料

 
    for item in data:                                             
        if item.find('h4').string != None :
            books = item.find('h4').string
        if item.find('li').a != None :
            autor = item.find('li').a.string
        if item.find('li', 'price_a').b != None :
            money = item.find('li', 'price_a').text
         
        writer.writerow([books, autor, money])          # 利用迴圈寫入多列資料
```

```python
#讀取csv檔案

import pandas as pd
datacsv = pd.read_csv("data.csv")
datacsv              
```

```python
#存為json檔

import json
datajson = datacsv.to_json(orient='records', force_ascii=False)

datajson
```
