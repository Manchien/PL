# [初任人員每人每月經常性薪資](https://data.gov.tw/dataset/6647)
### 各類初任人員每人每月平均經常性薪資
* 資料來源：[政府開放資料平台](https://data.gov.tw/)
* 「初任」，指由機關以外遴選具有法定任用資格之人員擔任定有官職等之職務

```python
#將JSON檔讀入Python

import json
f = open('MonthlySalaryByIndustry.json',encoding="utf8")
data2 = json.load(f)
type(data2)

#讀出來資料型態是list
```

```python
#把list轉成dict

data={}
for i in range(len(data2)):
    data[i]=data2[i]    
type(data)
```

```python
#讀一次資料

data
```

```python
print('總共有：',len(data),'種類的工作。')
```

```python
#第1種類工作

temp=data.get(1)
temp
```

```python
#讓使用者輸入查詢工作種類並印出

y = int(input("請輸入第幾種類的工作(0-115)："))
temp = data.get(y)

print("\n",temp)
```

```python
ask = temp.get("行業別")
print('行業別為：',ask)
```

```python
#印出此行業別的各職位薪水

if ask == temp.get("行業別"):
    
    print (ask, '：\n\n經常性薪資是', temp.get("經常性薪資-薪資"), '元，') 
    print ('專業人員薪資是', temp.get("專業人員-薪資"), '元，') 
    print ('技術員及助理專業人員薪資是', temp.get("技術員及助理專業人員-薪資"), '元。') 
    print ('事務支援人員薪資是', temp.get("事務支援人員-薪資"), '元，') 
    print ('服務及銷售工作人員薪資是', temp.get("服務及銷售工作人員-薪資"), '元。') 
    print ('技藝、機械設備操作及組裝人員薪資是', temp.get("技藝_機械設備操作及組裝人員-薪資"), '元，') 
    print ('基層技術工及勞力工薪資是', temp.get("基層技術工及勞力工-薪資"), '元。') 
    
    
```

```python
#將各職位薪水存入變數並進行排序

a1 = temp.get("經常性薪資-薪資")
a2 = temp.get("專業人員-薪資")
a3 = temp.get("技術員及助理專業人員-薪資")
a4 = temp.get("事務支援人員-薪資")
a5 = temp.get("服務及銷售工作人員-薪資")
a6 = temp.get("技藝_機械設備操作及組裝人員-薪資")
a7 = temp.get("基層技術工及勞力工-薪資") 
wage_list = [a1,a2,a3,a4,a5,a6,a7]
wage_index = ['經常性薪資','專業人員薪資','技術員及助理專業人員薪資',
         '事務支援人員薪資','服務及銷售工作人員薪資',
        '技藝、機械設備操作及組裝人員薪資','基層技術工及勞力工薪資']

#zip()函數：將兩個序列合并，可强制轉換為列表或字典
wage = zip(wage_index,wage_list)
list_temp = list(wage)

#list排序：使用lambda函式來取出每個元素的第[1]個元素來比較
list_temp = sorted(list_temp,key=lambda x: x[1])
                                    # x為list內每個元素，依照x[1]的元素進行比較

print(list_temp)
```

```python
## 通過此程式，我們可知道初任人員的薪水排序。
```
