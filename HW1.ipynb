## `109年設有癌症資源中心的醫院` 與 `109年實施「全方位癌症防治策進計畫」的醫院`
### 國健署提供的免費癌症篩檢服務，包括：
* 乳癌：45-69歲婦女、40-44歲二等血親內曾罹患乳癌之婦女，每2年1次乳房攝影檢查。
* 大腸癌：50-74歲民眾，每2年1次糞便潛血檢查。
* 子宮頸癌：30歲以上婦女，每3年應至少接受1次子宮頸抹片檢查。
* 口腔癌：30歲以上有嚼檳榔（含已戒檳榔）或吸菸之民眾、18歲以上有嚼檳榔（含已戒檳榔）之原住民，每2年1次口腔黏膜檢查。
* 肺癌：(1)具肺癌家族史：50至74歲男性或45至74歲女性，且其父母、子女或兄弟姊妹經診斷為肺癌之民眾。(2)重度吸菸史：50至74歲吸菸史達30包-年以上，有意願戒菸或戒菸15年內之重度吸菸者，每2年1次低劑量電腦斷層檢查(LDCT)。

#### 這兩份醫院名單可以讓我找出有哪些醫院是在我做完癌症篩檢之後，如果是需要治療的狀況，可以直接掛號而不需要另外找其他醫院。

```python
import pandas as pd

# read csv file
cpp = pd.read_csv("CancerPreventionProgram.csv")#109年執行「全方位癌症防治策進計畫」的醫院
cr = pd.read_csv("CancerResources.csv")# 109年設有癌症資源中心的醫院 
```

```python
cpp.tail()#109年執行「全方位癌症防治策進計畫」的醫院
```

```python
cr.tail()# 109年設有癌症資源中心的醫院 
```

```python
alldatacr = list(cr['醫院名稱'])
alldatacpp = list(cpp['機構名稱'])
print ("109年設有癌症資源中心的醫院數：", len(alldatacr), " vs. 109年「全方位癌症防治策進計畫」的醫院數：", len(alldatacpp))
```
